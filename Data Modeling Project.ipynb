```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    #connecting to default PostgreSQL database
    conn = psycopg2.connect("host=127.0.0.1 dbname=postgres user=postgres password=postgresql")
    #Set automatic commit to be true so that each action is committed without having to call conn.commit() after each command
    conn.set_session(autocommit=True)
    #Use the connection to get a cursor that can be used to execute queries
    cur = conn.cursor()
    
    #create database accounts with UTF-8 encoding
    cur.execute("DROP DATABASE IF EXISTS accdb")
    cur.execute("CREATE DATABASE accdb")
    
    #close connection to default database
    conn.close()
    
    #connect to accounts database
    conn = psycopg2.connect("host=127.0.0.1 dbname=accdb user=postgres password=postgresql") #connection with the database
    cur = conn.cursor()    #for execution of teh queries
    
    return cur, conn
```
